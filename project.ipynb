## API

```python
#public API from Brewdog brewery with recipes for 24 beers

url = 'https://api.punkapi.com/v2/beers'

response = requests.get(url)

results = response.json()
results
```

```python
#3 functions to extract information about the 3 main ingredients for each beer
#malts, hops and yeast

#we search name of beer, name of malt, and amount

def get_malts():
    for beer in results:
        ingredients = beer['ingredients']
        malts = ingredients['malt']
        for malt in malts:
            malt_name = malt['name'] 
            malt_names.append(malt_name)
            malt_amount = malt['amount']['value']
            malt_amounts.append(malt_amount)
            beer_name = beer['name']
            beer_names_m.append(beer_name)
            
#we search name of beer, name of hop, and amount
            
def get_hops():
    for beer in results:
        ingredients = beer['ingredients']
        hops = ingredients['hops']
        for hop in hops:
            hop_name = hop['name'] 
            hop_names.append(hop_name)
            hop_amount = hop['amount']['value']
            hop_amounts.append(hop_amount)
            beer_name = beer['name']
            beer_names_h.append(beer_name)
            
#we search name of beer and name of yeast            

def get_yeast():
    for beer in results:
        ingredients = beer['ingredients']
        yeasts = ingredients['yeast']
        yeast_names.append(yeasts)
        beer_name = beer['name']
        beer_names_y.append(beer_name)
    
```

```python
#a new list with ingredients for each beer will be created in these lists

malt_names = []
malt_amounts = []
hop_names = []
hop_amounts = []
yeast_names = []
beer_names_m = []
beer_names_h = []
beer_names_y = []


#run functions to get info from JSON

get_malts()
get_hops()
get_yeast()
```

```python
#creation of dataframes for each ingredient

beer_recipes_malt = pd.DataFrame()
beer_recipes_malt['beer_name'] = beer_names_m
beer_recipes_malt['malts'] = malt_names
beer_recipes_malt['malt_amount_kg'] = malt_amounts

beer_recipes_hop = pd.DataFrame()
beer_recipes_hop['beer_name'] = beer_names_h
beer_recipes_hop['hops'] = hop_names
beer_recipes_hop['hops_amount_g'] = hop_amounts

beer_recipes_yeast = pd.DataFrame()
beer_recipes_yeast['beer_name'] = beer_names_y
beer_recipes_yeast['yeast'] = yeast_names
```

## Web scrapping

```python
#ratebeer contains information and ratings for beers
#this url contains all beers from Brewdog brewery

url_scrap = 'https://www.ratebeer.com/Ratings/Beer/ShowBrewerBeers.asp?BrewerID=8534'
```

```python
#I will use a random user agent every time I make a request
#user agents list is in a txt file called `user_agents.txt`

import random
with open('user_agents.txt') as f:
    lines = f.readlines()
    lst = [line.strip() for line in lines[0::2]]
    random_ua = random.choice(lst)
      
headers = {'user-agent': random_ua}
```

```python
#request to get html from the url

html = requests.get(url_scrap, headers).content
html[:500]
```

```python
soup = BeautifulSoup(html, 'lxml')
soup
```

### Find beer names

```python
#there are 695 rows of beers and one with the variable names

beers = soup.find_all('tr')
print(len(beers))
beers
```

```python
#Finding beers names

beers2 = [beer.find('strong') for beer in beers[1:]] #[1:] to discard first row with variable names
beers2
```

```python
beers_list = [beer.text.strip() for beer in beers2]
beers_list
```

```python
#Deleting Brewdog and / at the start of names
#first, split names

beers_list2 = [beer.split() for beer in beers_list]
beers_list2
```

```python
#and delete

for beer in beers_list2:
 
    if "BrewDog" in str(beer):
        beer.remove("BrewDog")
    if "/ " in str(beer):
        beer.remove("/ ")
        
beers_list2
```

```python
#and rejoin names

beers_list3 = [' '.join(beer) for beer in beers_list2]
beers_list3
```

```python
len(beers_list3)
```

### Find styles

```python
styles = soup.find_all('span', {'class':'real-small hidden-xs'})
styles
```

```python
style_list = [style.text.strip() for style in styles]
style_list
```

```python
len(style_list)
```

```python
#creating a new dataframe joining each name with its style

df_beer_styles = pd.DataFrame()
df_beer_styles['beer_name'] = beers_list3
df_beer_styles['style'] = style_list

df_beer_styles.head(25)
```

## Merging dataframes from API and scrapping

```python
#joining tables to associate each beer name with its style

beer_both = recipes_final.join(df_beer_styles.set_index('beer_name'), on='beer_name')
beer_both.head(50)
```
